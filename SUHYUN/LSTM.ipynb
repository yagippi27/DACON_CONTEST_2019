```python
print(type(X_train))
print(X_train.shape)
X_train = X_train.values
print(type(y_train))
print(y_train.shape)
y_train = y_train.values
```

```python
X_train_t = X_train.reshape(X_train.shape[0], 3708, 1)
print("최종 DATA")
print(X_train_t.shape)
print(X_train_t)
print(y_train)
```

```python
from keras.layers import LSTM 
from keras.models import Sequential 
from keras.layers import Dense 
import keras.backend as K 
from keras.callbacks import EarlyStopping

K.clear_session()
    
model = Sequential() # Sequeatial Model 
model.add(LSTM(20, input_shape=(3708, 1))) # (timestep, feature) 
model.add(Dense(3708)) # output = 1 
model.compile(loss='mean_squared_error', optimizer='adam') 
model.summary()
```

```python
early_stop = EarlyStopping(monitor='loss', patience=10)

model.fit(X_train_t, y_train, epochs=100, batch_size=30, callbacks=[early_stop])
```
