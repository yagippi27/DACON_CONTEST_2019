```python
train = pd.read_csv("./DACON_DATASET/revise_dataset/train_DACON_NA_remove.csv")
r48_med = pd.read_csv('SUHYUN/rolling/train_NA/r48/r48_med.csv')
```

```python
total_nan = []
replace_nan = []
for k in range(1, len(train.columns)):
    total_nan.append(train.iloc[:, k].isna().sum())
    n = 0
    for i in range(len(train.index)):
        if math.isnan(train.iloc[i, k]) == True:
            train.iloc[i,k] = r48_med.iloc[i,k]
            if math.isnan(train.iloc[i,k]) == False:
                n += 1
        else:
            pass
    replace_nan.append(n)    
    print(k, "열,", i, "행, 대체 nan 수:", n)
```

```python
train.to_csv("./DACON_DATASET/revise_dataset/train_DACON_NA_remove_and_r48_med.csv", index = False)
```

```python
replace_rato = []
for i in range(len(total_nan)):
    if total_nan[i] != 0:
        replace_rato.append(replace_nan[i]/total_nan[i])
    else:
        replace_rato.append(replace_nan[i]/1)
    print(total_nan[i], replace_nan[i],replace_rato[i])
```

```python
data = {"total_nan": total_nan, "replace_nan": replace_nan, "replace_rato": replace_rato}
df = pd.DataFrame(data, columns = ['total_nan', 'replace_nan', 'replace_rato'])
```
